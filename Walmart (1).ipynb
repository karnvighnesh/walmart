```python
import scipy.stats as stats
def confidence_Interval(sample_data,confidence=0.95):
  sample_mean=np.mean(sample_data)
  sample_std=np.std(sample_data,ddof=1)
  sample_size=len(sample_data)
  se= sample_std/np.sqrt(sample_size)

  z_value=stats.norm.ppf(1-((1-confidence)/2))
  lower_bound=sample_mean-z_value*se
  upper_bound=sample_mean+z_value*se
  return sample_mean, (lower_bound, upper_bound)
```

```python
male_spending = data[data['Gender'] == 'M']['Purchase']
female_spending = data[data['Gender'] == 'F']['Purchase']
```

```python
confidence_interval=[0.90,0.95,0.99]
for conf in confidence_interval:
  male_mean, male_ci = confidence_Interval(male_spending, confidence=conf)
  female_mean, female_ci = confidence_Interval(female_spending, confidence=conf)
  print(f"\nFor {conf*100}% Confidence Level:")
  print(f"Male Spending -> Mean: {male_mean:.2f}, Interval: {male_ci}")
  print(f"Female Spending -> Mean: {female_mean:.2f}, Interval: {female_ci}")
  print(f"Overlapping: {'Yes' if male_ci[0] <= female_ci[1] and female_ci[0] <= male_ci[1] else 'No'}")
```

```python
married_spending = data[data['Marital_Status'] == 1]['Purchase']
unmarried_spending = data[data['Marital_Status'] == 0]['Purchase']

print("\nConfidence Intervals for Married vs Unmarried:")
for conf in confidence_interval:
    married_mean, married_ci = confidence_Interval(married_spending, confidence=conf)
    unmarried_mean, unmarried_ci = confidence_Interval(unmarried_spending, confidence=conf)

    print(f"\nFor {conf*100}% Confidence Level:")
    print(f"Married Spending -> Mean: {married_mean:.2f}, Interval: {married_ci}")
    print(f"Unmarried Spending -> Mean: {unmarried_mean:.2f}, Interval: {unmarried_ci}")
    print(f"Overlapping: {'Yes' if married_ci[0] <= unmarried_ci[1] and unmarried_ci[0] <= married_ci[1] else 'No'}")
```

Comments on the Boxplot of Purchase Amounts by Age Groups

1. Median Purchase Amount
The black horizontal line inside each box represents the median purchase amount for each age group.

The median values across all age groups seem to be relatively similar, suggesting that age does not drastically impact the typical purchase amount.

2. Spread of Purchase Amounts

The boxes represent the interquartile range (IQR) (25th to 75th percentile).
The purchase amount distribution is consistent across all age groups, with similar IQR widths.

This suggests that the spending habits of different age groups are not significantly different in terms of variance.

3. Outliers

The dots above the whiskers represent outliers—transactions that are significantly higher than the majority of purchases.

Every age group has similar outliers, indicating that high-value purchases occur across all age groups, not just in a particular category.

4. Whiskers (Range of Data)

The whiskers (vertical lines extending from the boxes) indicate the range within 1.5 times the IQR.
The overall range of purchase amounts is similar across all age groups, suggesting no strong age-related impact on spending.
